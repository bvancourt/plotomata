# Plotomata Usage Demonstration

```python
import os
import numpy as np
import pandas as pd
import plotomata

# for debugging
import importlib
```

```python
# load some test data
test_data_1 = pd.read_excel("./test_data/test_data_1.xlsx")
test_data_1.head(17)
```

```python
# Scatter plot with size and color legends
importlib.reload(plotomata)
plotomata.scatter_plot(
    test_data_1["Floats X"],
    test_data_1["Floats Y"],
    test_data_1["Strings C"],
    mode="categorical",
    size=test_data_1["Ints S"],
    size_to_area_func=lambda s: np.array(s+.05, dtype=np.float64)**(1/5),
    title="Scatter Plot",
    size_legend_title="Ints S",
    color_legend_title="Strings C",
    x_label="Floats X",
    y_label="Floats Y",
    color_palette={
        "category 1" : (0, 0, 0, 1), # black
    #   "category 2" : colors for missing labes are taken from default color palette.
        "category 3" : (1, 0, 0, 1), # red
    }
)
```

```python
# Scatter plot with size and colorbar
importlib.reload(plotomata)
plotomata.scatter_plot(
    {
        "x" : test_data_1["Floats X"],
        "y" : test_data_1["Floats Y"],
        "c" : test_data_1["Ints S"],
    },
    size=10/(2+np.log(test_data_1["Ints S"])),
    mode="color_map",
    title="Scatter Plot",
    colorbar_label="Ints S",
    color_legend_title="Strings C",
    x_label="Floats X",
    y_label="Floats Y",
    cmap_norm="log",
    size_legend_title="Dot Size",
)
```
